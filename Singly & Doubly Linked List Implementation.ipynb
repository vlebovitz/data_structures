# Singly Linked List Implementation

```python
class Node:
    def __init__(self,value,next=None):
        self.value = value
        self.next = next
class LinkedList:
    def __init__(self,value=None):
        self.value = value
        self.head = Node(self.value)
    def append(self,new_value):
        
        node = Node(new_value)
        
        if self.head is None:
            self.head = node
            return
        curr = self.head
        while(curr.next):
            curr = curr.next
        curr.next = node
        return
    
    def prepend(self,new_value):
        node = Node(new_value)
        
        if self.head is None:
            self.head = node
            return
        node.next = self.head
        self.head = node
        return
    
    def stringify(self):
        result = ""
        curr = self.head
        while(curr):
            result += str(curr.value) + " -> "
            curr = curr.next
        return result
```

# Doubly Linked List Implementation

```python
class Node:
    def __init__(self,value,next=None,prev=None):
        self.value = value
        self.next = next
        self.prev = prev
class DoublyLinkedList(LinkedList):
    def __init__(self,value=None):
        self.value = value
        if self.value is not None:
            self.head = self.tail = Node(self.value)
            self.length = 1
        else:
            self.head = self.tail = None
            self.length = 0
    
    def prepend(self,new_value):
        self.length +=1
        new_node = Node(new_value)
        if self.head is None:
            self.head = self.tail = new_node
            return
        new_node.next = self.head
        self.head.prev = new_node
        self.head = new_node
    
    def append(self, new_value):
        self.length += 1
        new_node = Node(new_value)
        if self.head is None:
            self.head = self.tail = new_node
            return
        self.tail.next = new_node
        new_node.prev = self.tail
        self.tail = new_node
    
    def removeAt(self,idx):
        if (idx > self.length):
            print("Must Enter an In Bounds Value")
            return
        self.length -= 1
        if (idx == 1 and self.head is not None):
            deleteNode = self.head
            self.head = self.head.next
            deleteNode = None
            if (self.head != None):
                self.head.prev = None
            return
        curr = self.head
        for i in range(1, idx):
            curr = curr.next
        
        if (curr.prev):
            curr.prev = curr.next
        if (curr.next):
            curr.next = curr.prev
        
        
            
        
            
            
            
            
    
    def stringify(self):
        if self.head is None:
            return
        curr = self.head
        res = ""
        while(curr):
            res += " <- " + str(curr.value) + " -> "
            curr = curr.next
        return res  
```
