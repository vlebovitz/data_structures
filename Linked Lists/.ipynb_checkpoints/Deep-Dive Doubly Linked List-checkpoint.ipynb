```python
#Implement A Doubly Linked List
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None
        self.prev = None
class DoublyLinkedList:
    def __init__(self,value=None):
        if value == None:
            self.length = 0
            self.head = self.tail = None
        else:
            new_node = Node(value)
            self.head = self.tail = new_node
            self.length = 1
    def print_list(self):
        if self.length < 1:
            return
        curr = self.head
        while curr:
            print(curr.value)
            curr = curr.next
    def append(self,new_value):
        new_node = Node(new_value)
        if self.length < 1:
            self.length += 1
            self.head = self.tail = new_node
            return
        self.length += 1
        self.tail.next = new_node
        new_node.prev = self.tail
        self.tail = new_node
        return True
    def pop(self):
        if self.length < 1:
            return
        if self.length == 1:
            self.head = self.tail = None
        else:
            temp = self.tail
            self.tail = self.tail.prev
            self.tail.next = None
            temp.prev = None
        self.length -= 1
        return temp
    def prepend(self,new_value):
        new_node = Node(new_value)
        if self.length < 1:
            self.head = self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
        self.length += 1
        return True
    def pop_first(self):
        if self.length < 1:
            return
        temp = self.head
        if self.length == 1:
            self.head = self.tail = None
        else:
            self.head = self.head.next
            self.head.prev = None
            temp.next = None
        self.length -= 1
        return temp
    def get(self,index):
        if index < 1 or index > self.length:
            print("Out of Bounds")
            return
        curr = self.head
        for _ in range(index-1):
            curr = curr.next
        return curr
    def set_value(self,index,new_value):
        node_to_change = self.get(index)
        if node_to_change is None:
            return
        else:
            node_to_change.value = new_value
    def insert(self,index,new_value):
        if index < 1 or index > self.length:
            print("Out of Bounds")
            return
        
        if index == 1:
            self.prepend(new_value)
        else:
            new_node = Node(new_value)
            prev = None
            curr = self.head
            for i in range(index-1):
                prev = curr
                curr = curr.next
            new_node.next = curr
            new_node.prev = prev
            curr.prev = new_node
            prev.next = new_node
        self.length += 1
    def remove(self,index):
        if index < 1 or index > self.length:
            print("Out of Bounds")
            return
        if index == 1:
            self.pop_first()
        elif index == self.length:
            self.pop()
        else:
            prev = None
            curr = self.head
            for i in range(index-1):
                prev = curr
                curr = curr.next
            prev.next = curr.next
            curr.next.prev = prev
            curr.next = curr.prev = None
        
        
            
            
        
            

d1 = DoublyLinkedList()
d1.append(12)
d1.append(15)
d1.append(16)
d1.print_list()
d1.remove(2)
d1.print_list()
```
