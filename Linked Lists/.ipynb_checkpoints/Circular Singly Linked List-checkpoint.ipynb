```python
class Node:
    def __init__(self,value,next=None):
        self.value = value
        self.next = next
class CSLinkedList:
    def __init__(self,value=None):
        self.value = value
        if self.value is None:
            self.length = 0
            self.head = self.tail = None
        else:
            new_node = Node(self.value)
            new_node.next = new_node
            self.head = self.tail = new_node
            self.length = 1
            
    def append(self,value):
        new_node = Node(value)
        if self.length == 0:
            new_node.next = new_node
            self.head = self.tail = new_node
            self.length += 1
        else:
            self.tail.next = new_node
            self.tail = new_node
            new_node.next = self.head
        self.length += 1
        
    def prepend(self,value):
        new_node = Node(value)
        if self.length == 0:
            new_node.next = new_node
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.next = self.head
            self.head = new_node
        self.length += 1
    
    def insert(self,index,value):
        new_node = Node(value)
        if index > self.length or index < 0:
            print("Select New Index")
            return
        elif index == 0:
            self.tail.next = new_node
            new_node.next = self.head
            self.head = new_node
        elif index == self.length:
            self.tail.next = new_node
            self.tail = new_node
            new_node.next = self.head
        else:
            curr = self.head
            for i in range(index):
                prev = curr
                curr = curr.next
            new_node.next = curr
            prev.next = new_node
        self.length += 1
    
    def search(self,target):
        curr = self.head
        while curr.next:
            if curr.value == target:
                return True
            curr = curr.next
            if curr == self.head:
                break
        return False
    
    def remove(self,target):
        if self.length == 0:
            return
        
        
        curr = self.head
        prev = self.tail
        while curr:
            if curr.value == target:
                prev.next = curr.next
                curr.next = None
            prev = curr
            curr = curr.next
            if curr == self.head:
                break
                
    def __str__(self):
        res = ""
        if self.head is None:
            return
        curr = self.head
        while curr.next:
            res += str(curr.value) + " -> "
            curr = curr.next
            if curr == self.head:
                break
        return res
            
            
        
```

```python
csLinkedList = CSLinkedList(10)
csLinkedList.append(12)
csLinkedList.append(69)
csLinkedList.append(15)
print(csLinkedList)
csLinkedList.prepend(100)
print(csLinkedList)
csLinkedList.insert(0,45)
print(csLinkedList)
csLinkedList.remove(45)
print(csLinkedList)
```
