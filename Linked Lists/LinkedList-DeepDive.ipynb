```python
# Basic Build of LinkedList
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
class LinkedList:
    def __init__(self,data=None):
        self.head = self.tail = Node(data)
        self.length = 1
    def print_list(self):
        res = ""
        curr = self.head
        while curr:
            if curr.next is None:
                res += str(curr.data)
            else:
                res += str(curr.data) + " -> "
            curr = curr.next
        return res
    
    def append(self,new_data):
        new_node = Node(new_data)
        self.length += 1
        if self.head is None:
            self.head = self.tail = new_node
            return
        self.tail.next = new_node
        self.tail = new_node 
        return True
    
    def pop(self):
        if self.head is None:
            return
        if self.length == 1:
            self.length -= 1
            self.head = self.tail = None
            return
        self.length -= 1
        prev = None
        curr = self.head
        while curr.next:
            prev = curr
            curr = curr.next
        prev.next = None
        self.tail = prev
        return curr
    def prepend(self,new_value):
        if self.head is None:
            return
        self.length += 1
        new_node = Node(new_value)
        new_node.next = self.head
        self.head = new_node
    def pop_first(self):
        if self.length == 0:
            return
        if self.length == 1:
            self.length -= 1
            self.head = self.tail = None
        else:
            temp = self.head
            self.head = self.head.next
            return temp.data
    def get(self,index):
        if index < 1 and index > self.length:
            print("Error Out of Bounds")
            return
        curr = self.head
        for _ in range(index-1):
            curr = curr.next
        return curr
    def set_value(self,index,value):
        returned_node = self.get(index)
        if returned_node:
            returned_node.data = value
            return True
        return False
    def insert(self,index,value):
        if index < 1 or index > self.length:
            print("Out of Bounds Error")
            return
        if index == 1:
            self.prepend(value)
            return
        new_node = Node(value)
        self.length += 1
        prev = None
        curr = self.head
        for _ in range(index-1):
            prev = curr
            curr = curr.next
        new_node.next = curr
        prev.next = new_node
    def remove(self,index):
        if index < 1 or index > self.length:
            print("Out of Bounds Error")
            return
        if index == 1:
            self.pop_first()
            return
        prev = None
        curr = self.head
        for _ in range(index-1):
            prev = curr
            curr = curr.next
        prev.next = curr.next
        curr.next = None
    def reverse(self):
        if self.length == 0:
            return
        curr = self.head
        self.head = self.tail
        self.tail = curr
        prev = None
        #nxt = curr
        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
        
            
        
        
        
        
    
            
        
            
        
       
            

l1 = LinkedList(12)    
l1.append(14)
l1.append(18)
l1.append(15)
l1.append(100)
l1.print_list()
l1.insert(4,3)
l1.print_list()
l1.reverse()

l1.print_list()
```
