```python
#Creation of a Linked List

#Node Class
class Node:
    def __init__(self,value,next=None):
        self.value = value
        self.next = next
class LinkedList:
    def __init__(self,value=None):
        self.value = value
        if self.value is None:
            self.length = 0
            self.head = self.tail = None
        else:
            self.length = 1
            self.head = self.tail = Node(self.value)
    
    def prepend(self,new_node):
        new_node = Node(new_node)
        self.length += 1
        if self.head is None:
            self.head = self.tail = new_node
            return
        new_node.next = self.head
        self.head = new_node
    
    def append(self,new_node):
        new_node = Node(new_node)
        self.length += 1
        if self.head is None:
            self.head = self.tail = new_node
            return
        self.tail.next = new_node
        self.tail = new_node
    
    def __str__(self):
        res = ""
        curr = self.head
        while curr:
            res += str(curr.value)+ " -> "  
            curr = curr.next
        return res
    
    def insert(self,new_node,index):
        if index > self.length or index < 0:
            print("Index Out Of Bounds!")
            return
        elif index == 0:
            self.length += 1
            self.prepend(new_node)
            return
        new_node = Node(new_node)
        self.length += 1
        curr = self.head
        for i in range(index-1):
            curr = curr.next
        new_node.next = curr.next
        curr.next = new_node
    
    def search(self,value):
        if self.head is None:
            return 
        curr = self.head
        while curr:
            if curr.value == value:
                return True
            curr = curr.next
        return False
    
    def get(self,index):
        if index == -1:
            return self.tail
        if index == 0:
            return self.head
        if index < 0 or index >= self.length:
            return None
        curr = self.head
        for i in range(index):
            curr = curr.next
        return curr
    
    def set(self,index,new_value):
        curr = self.get(index)
        if curr:
            curr.value = new_value
            return True
        return False
    
    def pop_first(self):
        curr = self.head
        self.head = self.head.next
        curr.next = None
        self.length -= 1
        return curr.value
    
    def pop(self):
        curr = self.head
        for i in range(1,self.length-2):
            #print(curr.value,end="\n")
            curr =  curr.next
        self.tail = curr
        curr.next = None
    
    def remove(self,index):
        curr = self.head
        for i in range(index):
            prev = curr
            curr = curr.next
        prev.next = curr.next
        curr.next = None
    
    def delete_all(self):
        self.head = None
        self.tail = None
        self.length = 0
        
        
        
            
            
            
            
        
l1 = LinkedList(12)
l1.append(14)
l1.prepend(35)
l1.append(23)
l1.append(50)
#print(l1)

l1.insert(69,0)
#print(l1.search(121))
print(l1)
l1.remove(2)
print(l1)
l1.remove(4)
print(l1)

```
